# Movie Recommender System

```python
# Importing Library
import pandas as pd
import numpy as np
```

```python
# Loading dataset credits
df_credits=pd.read_csv("tmdb_5000_credits.csv")
df_credits
```

```python
# Loading dataset movies
df_movies=pd.read_csv("tmdb_5000_movies.csv")
df_movies
```

```python
# Merge both dataframes based on movie_id and title
df=df_movies.merge(df_credits,on='title')
df.info()
```

```python
# Dropping not used columns: 
# budget, homepage, original_language, original_title,
# popularity, production_companies, production_countries,
# release_date, revenue, runtime, spoken_languages, status,
# tagline, vote_average, vote_count, movie_id
df.drop(columns=['budget', 'homepage', 'original_language', 'original_title', 'popularity',
                'production_companies', 'production_countries', 'release_date', 'revenue',
                'runtime', 'spoken_languages', 'status', 'tagline', 'vote_average', 'vote_count', 'movie_id'],
       inplace=True)
df.head()
```

```python
# Checking for null
df.isnull().sum()
```

```python
# Dropping null values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
# Check for duplicate
df.duplicated().sum()
```

```python
# Preprocess genres
import ast # convert string dict to dict
def convert_genres(obj_string_list):
    l=[]
    for i in ast.literal_eval(obj_string_list):
        l.append(i['name'])
    return l    
```

```python
df['genres']=df['genres'].apply(convert_genres)
```

```python
df.head()
```

```python
# Preprocess keywords
df['keywords']=df['keywords'].apply(convert_genres)
```

```python
df.head()
```

```python
# Preprocess cast
# Get 1st, 5 Cast
import ast # convert string dict to dict
def convert_cast(obj_string_list):
    l=[]
    counter = 0
    for i in ast.literal_eval(obj_string_list):
        if(counter != 5):
            l.append(i['name'])
            counter=counter+1
        else:
            break
    return l   
```

```python
df['cast']=df['cast'].apply(convert_cast)
```

```python
df.head()
```

```python
# Preprocess crew
# Get Director name
import ast # convert string dict to dict
def convert_crew(obj_string_list):
    l=[]
    for i in ast.literal_eval(obj_string_list):
        if(i['job'] == 'Director'):
            l.append(i['name'])
            break
    return l  
```

```python
df['crew']=df['crew'].apply(convert_crew)
```

```python
df.head()
```

```python
# Preprocess overview
# convert to list
df['overview']=df['overview'].apply(lambda x:x.split())
```

```python
df.head()
```

```python
# Remove spaces between words
df['genres']=df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['keywords']=df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast']=df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew']=df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df.head()
```

```python
df['tags']=df['overview']+df['genres']+df['keywords']+df['cast']+df['crew']
```

```python
df.drop(columns=['genres', 'keywords', 'overview', 'cast', 'crew'], inplace=True)
```

```python
df.head()
```

```python
# Convert list in tags to string
df['tags']=df['tags'].apply(lambda x: " ".join(x))
```

```python
df.head()
```

```python
# convert to lower case
df['tags']=df['tags'].apply(lambda x: x.lower())
df['title']=df['title'].apply(lambda x: x.lower())
```

```python
df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def do_stem(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)     
```

```python
df['tags']=df['tags'].apply(do_stem)
```

## Text Vectorization using Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words='english', max_features=5000)
vectors=cv.fit_transform(df['tags']).toarray()
```

```python
# cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
def recommend_movies(movie_name):
    try:
        movie_index=df[df['title']==movie_name.lower()].index[0]
        distances=similarity[movie_index]
        # sort in descending and get top 5 movies
        movie_list=sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

        for i in movie_list:
            print(f"Recommended Movie {df.iloc[i[0]].title} with ID={df.iloc[i[0]].id}")
    except IndexError:
        print("Not in Database")
    
```

```python
recommend_movies('Avatssar')
```
